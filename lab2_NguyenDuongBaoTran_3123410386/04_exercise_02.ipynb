## Bài 1. Tính toán trên dãy số

```python
import numpy as np #Khai báo thư viện numpy

#1. Tạo một dãy số a có N phần tử (N = 10)
print("1. Tạo một dãy số a có N phần tử (N = 10)")
N = 10

a = np.random.randint(120, 191, N)
print("Day so a: ", a)

#2. Các thao tác trên dãy số a
print("\n2. Các thao tác tính toán đơn giản với dãy số a")

a2 = a**2
print("a. Binh phuong a: ", a2)

na = len(a)
print("\nb. Do dai a: ", na)

v_max = np.max(a)
v_min = np.min(a)
v_avg = np.mean(a)
print(f"\nc. Gia tri lon nhat: [{v_max}], gia tri nho nhat: [{v_min}], gia tri trung binh [{v_avg:.1f}]")

v_sigma2 = np.var(a, ddof=1)  # phương sai với n-1
v_var = np.sqrt(v_sigma2)     # độ lệch chuẩn
print(f"\nd. Phuong sai: [{v_sigma2:.2f}] va do lech chuan: [{v_var:.2f}]")

#3. Xử lý khác trên dãy số
print("\n3. Xử lý khác trên dãy số")

a = np.array([120, 183, 160, 175, 145, 162, 190, 160, 152, 162])

#a. Tạo dãy số x
x = np.arange(1, 11)
print(f"a. Dãy số x: {x}")

#b. Liệt kê và đếm số phần tử chẵn
a_la_chan = (a % 2 == 0)           
a_chan = a[a_la_chan]              
n_a_chan = int(a_la_chan.sum())    

x_la_chan = (x % 2 == 0)           
x_chan = x[x_la_chan]              
n_x_chan = int(x_la_chan.sum())    

print( "b. Cac phan tu chan trong day a, b: ")
print(f"+ Day a: {a}")
print(f"  - Loc chan: {a_la_chan}")
print(f"  - Day chua phan tu chan trong a: {a_chan}, so luong {n_a_chan} phan tu.")
print(f"+ Day x: {x}")
print(f"  - Loc chan: {x_la_chan}")
print(f"  - Day chua phan tu chan trong a: {x_chan}, so luong {n_x_chan} phan tu.")

#c. Tính khoảng cách ở vị trí lẻ
v_le = [1, 3, 5, 7, 9]    
a_le = a[v_le]
x_le = x[v_le]
hieu_le = a_le - x_le
kc_le = np.sqrt(np.sum(hieu_le ** 2))

print( "\nc.Tính hiệu các phần tử ở vị trí lẻ của mảng a và x: ")
print(f"+ Vi tri le trong a va x: {v_le}")
print(f"+ Day so a: {a} va cac so o vi tri le: {a_le}")
print(f"+ Day so x: {x} va cac so o vi tri le: {x_le}")
print(f"+ Hieu cua a va x o vi tri le: {hieu_le}")
print(f"+ Khoang cach cac phan tu cua a va x o vi tri le: {kc_le: .2f}")

#d. Tìm khoảng cách nhỏ nhất giữa tập a và x
kc = np.abs(a[:, np.newaxis] - x)  # ma trận chênh lệch tuyệt đối
v_min = np.min(kc)

print("\nd. Tìm khoảng cách nhỏ nhất giữa tập a và x:")
print(f"+ Day a: {a}")
print(f"+ Day x: {x}")
print(f"+ Khoang cach nho nhat: {v_min}")

#4. Sinh ngẫu nhiên dãy số
print("\n4. Sinh ngẫu nhiên dãy số")

import matplotlib.pyplot as plt

a = np.random.randint(1, 11, size=10000)   
value, cnt = np.unique(a, return_counts=True)

print("a. Sinh ngẫu nhiên theo phân bố đều")
print(f"+ Gia tri: {value}")
print(f"+ So lan xuat hien: {cnt}")
print("+ Do thi ham so xuat hien")
plt.figure(figsize=(6,6))      #  kich thuoc ban ve
plt.bar(value, cnt)            #  ve cac khoi chu nhat theo day x, y
plt.xlim(0, 11)                #  gioi han truc x
plt.ylim(0, np.max(cnt) + 10)  #  gioi han truc y
plt.savefig("4a.png")          #  luu do thi
plt.show()                     #  hien thi do thi

mu, sigma = 1.0, 1.0
N = 500
x_min, x_max = mu - 5*sigma, mu + 5*sigma
x = np.random.rand(N) * (x_max - x_min) + x_min   # random đều trong đoạn [x_min, x_max]
y = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-((x - mu)**2)/(2*sigma**2))

print("b. Sinh cặp (x, y) với y làm hàm phân phối chuẩn")
print(f"+ x: {x_min} <= min [{np.min(x): .2f}] <= max [{np.max(x): .2f}] <= {x_max}")
print(f"+ Ve ham so y(x)")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.scatter(x, y)              # ve cac diem (xi, yi)
plt.savefig("4b.png")          #  luu do thi
plt.show()                     #  hien thi do thi

N = 100000
mu, sigma = 0, 1
x = mu + sigma * np.random.randn(N)   # sinh N số theo N(mu, sigma^2)
cnt, val = np.histogram(x, bins=300, density=True)
val = (val[:-1] + val[1:]) / 2        # lấy trung điểm để vẽ

print("\nc. Sinh day so theo phan phoi chuan")
print(f"+ x: len = {len(x)}, min [{np.min(x)}], max [{np.max(x)}]")
print(f"+ Mat so xac suat cua day so x")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.xlim(np.min(val), np.max(val))
plt.scatter(val, cnt)
plt.show()
```

## Bài 2. Tính toán trên ma trận

```python
import numpy as np

#1. Tạo 1 ma trận a có M = 6 dòng, N = 4 cột
M, N = 6, 4
a = np.random.randint(100, 200, (M, N))

print("1. Tạo ma trận a có M = 6 dòng, N = 4 cột")
print("Ma trận a: \n", a)

#2. Các thao tác trên ma trận a
print("\n2. Các thao tác trên ma trận a")

# a. Ma trận chuyển vị
a_t = a.T

# b. Phần tử dòng 2 cột 3
v_23 = a[2, 3]

# c. Trích xuất dòng
d_0    = a[0]      # dòng đầu
d_last = a[-1]     # dòng cuối
d_2    = a[2]      # dòng 2

# d. Trích xuất cột
c_l1 = a[:, -2]    # cột kế cuối
c_0  = a[:, 0]     # cột đầu

# e. Đảo các giá trị trên từng cột
dao_cot = np.flipud(a)

# f. Tính toán
tong_dong = np.sum(a, axis=1)     
tb_cot    = np.mean(a, axis=0)   

print(f"a. In ra ma trận chuyển vị: \n{a_t}")
print(f"\nb. In ra phần tử ở vị trí dòng 2 cột 3: {v_23}")
print(f"\nc. Trích xuất các dòng: ")
print(f"   + Dòng đầu : {d_0}")
print(f"   + Dòng cuối: {d_last}")
print(f"   + Dòng hai : {d_2}")
print(f"\nd. Trích xuất các cột: ")
print(f"   + Cột kế cuối: {c_l1}")
print(f"   + Cột đầu: {c_0}")
print(f"\ne. Đảo các giá trị trên từng cột: \n {dao_cot}")
print(f"\nf. Phép tính theo trục:")
print(f"   + Tổng dòng: {tong_dong}")
print(f"   + Trung bình cột: {tb_cot}")

#3. Các toán tử trên hai ma trận
print("\n3. Các toán tử trên hai ma trận")
# a. Tạo ma trận A, B (4x3) và X (3x4)
A = np.random.randint(1, 6, (4, 3))
B = np.random.randint(1, 6, (4, 3))
X = np.random.randint(1, 6, (3, 4))

# b. Tính toán từng phần tử
AB_add = A + B
AB_sub = A - B
AB_mul = A * B  

# c. Tính phép nhân ma trận
Y = np.dot(A, X)

# In kết quả
print("a. Tạo hai ma trận A và B có 4 dòng x 3 cột và X có 3 dòng x 4 cột")
print(f"+ Ma tran A: \n{A}")
print(f"+ Ma tran B: \n{B}")
print(f"+ Ma tran X: \n{X}")

print("\nb. Tính tổng, hiệu, tích của từng phần tử trên hai ma trận A và B")
print(f"A + B = \n{AB_add}")
print(f"A - B = \n{AB_sub}")
print(f"A . B = \n{AB_mul}")

print("\nc. Tính phép nhân ma trận Y = A x X")
print(f"Y = A x X --> shape A {A.shape} x shape X {X.shape} = shape Y {Y.shape} \n{Y}")

#4. Giải hệ phương trình tuyến tính
print("\n4. Giải hệ phương trình tuyến tính")
# Ma trận A và vector y
A = np.array([[4, 3, -5],
              [-2, -4, 5],
              [8, 8, 0]])
y = np.array([2, 5, -3])

x = np.linalg.solve(A, y)

yy = A @ x
diff = np.linalg.norm(yy - y)

print(f"+ Ma tran A: \n{A}")
print(f"+ Vector y: {y}")
print(f"+ Nghiem x: {x}")
print(f"+ Kiem tra: yy = Ax = {yy} ==> ||yy -y ||_2 = {diff: .2f}")
```

## Bài 3. Tính gần đúng số pi dùng phương pháp Monte Carlo

```python
import matplotlib.pyplot as plt # khai báo thư viện vẽ pyplot
import math

circle = plt.Circle((0, 0), 1, color='g') # tạo đối tượng vòng tròn

fig, ax = plt.subplots(figsize=(6, 6)) # lấy figure và vùng vẽ

plt.plot(0, 0, 'o', color=(0.9, 0.9, 1.0), alpha=0.8) # vẽ điểm tại tâm (0,0)
ax.add_patch(circle) 

x = 0.75; y = math.sqrt(1 - (x ** 2)) # (x = 0.75, y = căn bậc 2 của 1 - x^2)
plt.arrow(0, 0, x, y)

plt.xlim(-1, 1) 
plt.ylim(-1, 1) 

plt.savefig('pi.png') 
plt.show()            

import numpy as np

def calc_pi_monte_carlo(n = 100):
    # Sinh ngẫu nhiên n điểm (x, y) trong [-1, 1]
    x = np.random.rand(n) * 2 - 1
    y = np.random.rand(n) * 2 - 1
    
    # Đếm số điểm nằm trong hình tròn x^2 + y^2 <= 1
    inside_circle = np.sum(x**2 + y**2 <= 1)
    
    # Tính pi xấp xỉ
    pi = 4 * inside_circle / n
    return pi

import math

print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
```
